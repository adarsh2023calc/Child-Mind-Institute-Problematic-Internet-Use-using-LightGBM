```python
import pandas as pd
import numpy as np
import os
```

```python

def Training_data_generator():
    # Path to the parquet training directory
    
    series_train="/kaggle/input/child-mind-institute-problematic-internet-use/series_train.parquet"
    
    df_train=pd.read_csv("/kaggle/input/child-mind-institute-problematic-internet-use/train.csv")
    print(df_train.shape)
    X=[]
    Y=[]
    Z=[]
    enmo=[]
    light=[]
    anglez=[]
    steps=[]
    
    for id1 in df_train['id']:
        folder ="id="+ id1
        parquet_directory= os.path.join(series_train,folder)
        file= os.path.join(parquet_directory,"part-0.parquet")
        try:
            
            parq = pd.read_parquet(file)
            X.append(parq['X'].mean())
            Y.append(parq['Y'].mean())
            Z.append(parq['Z'].mean())
            enmo.append(parq['enmo'].mean())
            anglez.append(parq['anglez'].mean())
            steps.append(max(parq['step']))
            
            
        except:
            X.append(np.nan)
            Y.append(np.nan)
            Z.append(np.nan)
            enmo.append(np.nan)
            anglez.append(np.nan)
            steps.append(np.nan)
            
            
    
    df_train['avg_x'] = X
    df_train['avg_y'] = Y
    df_train['avg_z'] = Z
    df_train['avg_enmo']= enmo
    df_train['anglez']= anglez
    df_train['steps']=steps
    df_train['avg_x']=df_train['avg_x'].fillna(df_train['avg_x'].mean())
    df_train['avg_y']=df_train['avg_y'].fillna(df_train['avg_y'].mean())
    df_train['avg_z']=df_train['avg_z'].fillna(df_train['avg_z'].mean())
    df_train['avg_enmo']= df_train['avg_enmo'].fillna(df_train['avg_enmo'].mean())
    df_train['anglez']= df_train['anglez'].fillna(df_train['anglez'].mean())
    df_train['steps']=df_train['steps'].fillna(df_train['steps'].mean())
    
    return df_train
```
